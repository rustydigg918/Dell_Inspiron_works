```python
def hanoi(n, f, h, t):
    hanoi(n-1,f,t,h)
    move(f,t)
    hanoi(n-1,h,f,t)
    
```

```python
hanoi(4,'A','B','C')
```

**ok, so it's entering into negative values due to which, the stack space is getting exceeded**

```python
def hanoi(n, f, h, t):
    if n == 0:
        pass
    else:
        hanoi(n-1,f,t,h)
        move(f,t)
        hanoi(n-1,h,f,t)
```

```python
hanoi(12, 'A','B','C')
```
